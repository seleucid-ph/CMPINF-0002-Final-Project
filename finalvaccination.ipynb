```python
import pandas as pd

df = pd.read_csv("covidvaccination.tsv", sep="\t")

pt_df = df[df['neighborhood_municipality'].str.contains('Pittsburgh', case=False)]
vacc_df = pt_df['bivalent_booster']

print(f"Theres a total of {vacc_df.sum()} people vaccinated in all {vacc_df.count()} neighborhoods of Pittsburgh. Thats an average of {vacc_df.mean():.0f} people vaccinated per neighborhood.")
```

```python
import pandas as pd

df = pd.read_csv("covidvaccination.tsv", sep="\t")

pt_df = df[df['neighborhood_municipality'].str.contains('Pittsburgh', case=False)]
vacc_df = pt_df['bivalent_booster']

max_row = vacc_df.idxmax()

print(pt_df.loc[max_row])
print(f"The city with the highest record of vaccinations is Squirrel Hill South, with a number of 5347 vaccinations! Thats ~{vacc_df.max()/vacc_df.count():.0f} times more than the average!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('covidvaccination.tsv', sep='\t')
df_pittsburgh = df[df['neighborhood_municipality'].str.contains('Pittsburgh', case=False)]

plt.figure(figsize=(12, 6))

ax = df_pittsburgh.plot.bar(x='neighborhood_municipality', y='bivalent_booster', color='skyblue', legend=False)

plt.title('Vaccinations in Pittsburgh Neighborhoods', fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Number of Vaccinations', fontsize=12)

plt.xticks(rotation=90, fontsize=5)

plt.tight_layout()
plt.show()
```

With our data provided, Squirrel Hill South in Pittsburgh is the best neighborhood there is, with a whopping 5347 vaccinations out of the total 92 neighborhoods, so if you want to prevent having covid, going to Squirrel Hill would be your number one choice.

```python
#first I created the database delq, so that I could manipulate the data with PANDAS
import pandas as pd
delq = pd.read_csv("TaxDelinquencies.tsv", sep="\t", engine="python")
delq.head()
```

```python
#Here I filtered out all the instances of tax delinquencies from Pittsburgh's suburbs, by only showing entries in Pittsburgh
import pandas as pd
count_the_burgh = delq[delq['billing_city'].str.contains('Pittsburgh', case=False, na=False)]
count_the_burgh.head()
```

```python
#Next I used PANDAS to count each unique value and categorize each of these instances by neighborhood
neighborhood_instances = delq['neighborhood'].value_counts()
neighborhood_list = list(neighborhood_instances.items())
for neighborhood, count in neighborhood_list:
    print(f"{neighborhood},\n{count}")
```
